### _Let's fill-in the missing values of neighbourhood based on latitude and longitude_

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="App")
location = geolocator.geocode("Nurtepe")
location
```

```python
df.loc[df.neighbourhood.isna()]
```

```python
def get_neighbourhood_from_coordinates(lat, long):
    location = geolocator.reverse(str(lat)+','+str(long))
    return location.raw['address'].get('road', '')

temp = df.loc[df.neighbourhood.isna()].copy()
print(get_neighbourhood_from_coordinates(temp.iloc[0].lat, temp.iloc[0].long), end='\n\n')
print(get_neighbourhood_from_coordinates(45.72723, -70.12))
```

```python
idx = df.loc[df.neighbourhood.isna()].index
df.loc[idx]
```

```python
df.loc[idx, 'neighbourhood'] = df.loc[idx].apply(lambda x: get_neighbourhood_from_coordinates(x.lat, x.long), axis=1)
```

```python
df[df['neighbourhood'].isna()]
```

```python
df.loc[idx].head(2)
```

```python
df['neighbourhood'].value_counts()
```

### _Now let's fix the same for column 'neighbourhood_group'_

```python
def get_ngroup_from_coordinates(lat, long):
    location = geolocator.reverse(str(lat) + ','+ str(long))
    return location.raw['address'].get('suburb')
```

```python
# getting all records, where neighbourhood_group is NaN
idx = df.loc[df['neighbourhood_group'].isna()].index
df.loc[idx]
```

```python
df.loc[idx, 'neighbourhood_group'] = df.loc[idx].apply(lambda x: get_ngroup_from_coordinates(x.lat, x.long), axis=1)
```

```python
df['neighbourhood_group'] = df['neighbourhood_group'].replace('brookln', 'Brooklyn')
df['neighbourhood_group'].value_counts()
```

```python
df.loc[idx]
```

### _Let's see fix the records with NaN 'lat' or/and 'long' attributes based on their neighbourhood_ 

```python
def get_lat_from_location(loc):
    location = geolocator.geocode(loc)
    return location.latitude
def get_long_from_location(loc):
    location = geolocator.geocode(loc)
    return location.longitude
```

```python
idx = df.loc[df['lat'].isna() & df['long'].isna()].index
df.loc[idx]
```

```python


df.loc[idx, 'lat'] = df.loc[idx].apply(
    lambda x: get_lat_from_location(x['neighbourhood'] + ', ' + x['neighbourhood_group']), axis=1
)

df.loc[idx, 'long'] = df.loc[idx].apply(
    lambda x: get_long_from_location(x['neighbourhood'] + ', ' + x['neighbourhood_group']), axis=1
)
```

```python
df.loc[idx]
```

### _Let's clean 'price' and 'service_fee' columns by removing the dollar sign and convert values to integers by fixing commas in numbers_

```python
df[['price']]
```

```python
def remove_currency(value):
    value = value.replace(',', '')
    value = value.replace('$', '')
    return value
print(remove_currency("$1,099"))
```

```python
df['price'] = df['price'].apply(lambda x: remove_currency(x))
df['price'] = df['price'].apply(int)

df['service_fee'] = df['service_fee'].apply(remove_currency)
df['service_fee'] = df['service_fee'].apply(int)
```

```python
print(df[['price']].value_counts(), end='\n\n')
print(df[['service_fee']].value_counts())
```
